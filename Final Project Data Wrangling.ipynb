# Covid - 19 Data Analysis Project using Python

```python
# Importing useful libraries
import numpy as np
import pandas as pd
```

# 1) Import Dataset from the URL

```python
#Importing dataset
url = 'https://raw.githubusercontent.com/SR1608/Datasets/main/covid-data.csv'
df = pd.read_csv(url)
df.head()
```

# 2) High level Data Understanding

```python
# a) Find no. of rows and columns in the dataset
print(f"Number of rows : {df.shape[0]}\nNumber of columns : {df.shape[1]}")
```

```python
# b) Find types of columns
df.dtypes
```

```python
# c) info and describe of data in dataframe
print("\n------------- Info of the dataset ----------------\n")
df.info()
print("\n\n---------------- Describe dataset ----------------\n")
df.describe()
```

# 3) Low level Data Understanding:

```python
# a) Find count of Unique values in location column
print("No. of unique values in location column : ",df.location.nunique())
```

```python
# b) Find which continent has maximum frquency using values
print("Continent having maximum frequency: ",df['continent'].value_counts().idxmax())
```

```python
# c) Find maximum and mean value in "total_cases".
print("Maximum value in 'total_cases' : ",df['total_cases'].max())
print("Mean value in 'total_cases' : ",df['total_cases'].mean())
```

```python
# d) Find 25%, 50%, 75% quartile value in 'total_deaths'
print(f"25% quartile value in 'total_deats' : {df.total_deaths.quantile(0.25)}\n\
50% quartile value in 'total_deats' : {df.total_deaths.quantile(0.5)}\n\
75% quartile value in 'total_deats' : {df.total_deaths.quantile(0.75)}")
```

```python
# e) Find which continent has maximum 'human_development_index'.
print("Continent with maximum 'human_development_index' : ",df.groupby('continent')['human_development_index'].max().idxmax())
```

```python
# f) Find which continent has minimum 'gdp_per_capita'.
print("Continent with minimum 'gdp_per_capita' : ",df.groupby('continent')['gdp_per_capita'].min().idxmin())
```

# 4) Filter the dataframe with only this columns:
['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index'] and update the data frame.

```python
df = df[['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index']]
```

# 5) Data Cleaaning:

```python
# a) Remove all duplicates observations
df.drop_duplicates(subset = None, keep = 'first', inplace = False)
```

```python
# b) Find missing values in all columns
df.isnull().sum()
```

```python
# c) Remove all observations where continent column value is missing
df = df.dropna(subset = ['continent'])
```

```python
# d. Fill all missing values with 0
df = df.fillna(0)
df
```

# 6) Date time format :

```python
# a) Convert date column in datetime format using pandas.to_datetime
df['date'] = pd.to_datetime(df['date'])
df.dtypes
```

```python
# b) Create new column month after extracting month data from date column.
df['Month'] = pd.DatetimeIndex(df['date']).month
df.head()
```

# 7) Data Aggregation

```python
# a) Find max value in all columns using groupby function on 'continent' column
# b) Store the result in a new dataframe named 'df_groupby'. (Use df_groupby dataframe for all further analysis)
df_groupby = df.groupby('continent').max().reset_index()
df_groupby
```

# 8) Feature Engineering :

```python
# a) Create a new feature 'total_deaths_to_total_cases' by ratio of 'total_deaths' column to 'total_cases'
df_groupby['total_deaths_to_total_cases'] = df_groupby['total_deaths'] / df_groupby['total_cases']
df_groupby
```

```python
# Matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
# seaborn
import seaborn as sns
```

# 9) Data Visualization :

```python
# a) Perform Univariate analysis on 'gdp_per_capita' column by plotting histogram using seaborn dist plot.
sns.distplot(df_groupby['gdp_per_capita'], kde = False, bins = 10)
```

```python
# b) Plot a scatter plot of 'total_cases' & 'gdp_per_capita'
plt.figure(figsize = (8,6))
x = df_groupby['total_cases']
y = df_groupby['gdp_per_capita']
plt.scatter(x, y, label = 'Points')
plt.xlabel("Total Cases")
plt.ylabel("GDP Per Capita")
plt.title("Scatter Plot")
plt.legend()
plt.show()
```

```python
# c) Plot Pairplot on df_groupby dataset.
sns.pairplot(df_groupby, palette = 'coolwarm')
```

```python
# d) Plot a bar plot of 'continent' column with 'total_cases'.
sns.barplot(x = 'total_cases', y = 'continent', data = df_groupby)
```
